```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('zillow_homes.csv')
```

```python
df.head()
```

```python
az_homes = df.loc[df.State.eq('AZ'),:]
```

```python
az_homes.head()
```

```python
def date_cleaner(month_string):
    yr = month_string[0:4]
    mnt = month_string[5:8]
    return yr + '-' + mnt + '-' + '01'

untouched = ['RegionName', 'State', 'CountyName']
time_periods_to_keep = list(az_homes.columns[6:])

tmp = (pd
 .melt(
#     az_homes.loc[az_homes.RegionName.eq('Mesa'),:], 
    az_homes,
    id_vars=untouched, 
    value_vars=time_periods_to_keep)
 .rename({'variable':'Month', 'value':'AverageHomePrice'}, axis=1)
)
tmp['Date'] = tmp.Month.apply(date_cleaner)
tmp.head(10)
```

```python
tmp.to_csv('clean_zillow_house_data.csv', index=False)
```
